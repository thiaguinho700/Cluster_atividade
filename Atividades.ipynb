```python
# Exercicio 4
# Importar as bibliotecas necess√°rias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Dados fornecidos
setores = [1, 2, 3, 4, 5, 6, 7, 8]
produtos_fabricados = [100, 50, 15, 200, 500, 1000, 375, 450]

# Criar um DataFrame com os dados
dados = pd.DataFrame({
    'Setor': setores,
    'Produtos Fabricados': produtos_fabricados
})

# Preparar os dados para o KMeans (usamos apenas a coluna 'Produtos Fabricados')
X = dados[['Produtos Fabricados']]

# Aplicar o algoritmo KMeans
n_clusters = 3  # N√∫mero de clusters desejados
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
dados['Cluster'] = kmeans.fit_predict(X)

# Visualizar os resultados
print("Dados com os clusters atribu√≠dos:")
print(dados)

# Gr√°fico para visualiza√ß√£o
plt.scatter(dados['Setor'], dados['Produtos Fabricados'], c=dados['Cluster'], cmap='viridis')
plt.title('Clusters dos setores mais produtivos')
plt.xlabel('Setor')
plt.ylabel('Produtos Fabricados')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Exercicio 5
# Importa√ß√£o das bibliotecas necess√°rias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Dados fornecidos
dias_semana = [1, 2, 3, 4, 5, 6, 7]  # 1 = Domingo, ..., 7 = S√°bado
numero_clientes = [15, 20, 30, 100, 350, 500, 700]

# Criando um DataFrame
dados = pd.DataFrame({
    'dia_semana': dias_semana,
    'numero_clientes': numero_clientes
})

# Prepara√ß√£o dos dados para o KMeans
X = dados[['dia_semana', 'numero_clientes']]

# Aplicando o algoritmo KMeans com um n√∫mero de clusters (k) definido
k = 3  # N√∫mero de clusters (alter√°vel conforme necessidade)
kmeans = KMeans(n_clusters=k, random_state=42)
dados['cluster'] = kmeans.fit_predict(X)

# Visualiza√ß√£o dos clusters
plt.figure(figsize=(10, 6))
for cluster in range(k):
    cluster_data = dados[dados['cluster'] == cluster]
    plt.scatter(cluster_data['dia_semana'], cluster_data['numero_clientes'], label=f'Cluster {cluster}')

# Configura√ß√µes do gr√°fico
plt.title('Clusters de Dias da Semana por N√∫mero de Clientes')
plt.xlabel('Dia da Semana (1 = Domingo, ..., 7 = S√°bado)')
plt.ylabel('N√∫mero de Clientes')
plt.legend()
plt.grid()
plt.show()

# Mostrando os resultados dos clusters
print("Resultados dos Clusters:")
print(dados)
```
